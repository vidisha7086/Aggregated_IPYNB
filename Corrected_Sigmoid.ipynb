```python
from sympy import exp, diff, symbols, simplify, lambdify
import numpy as np
import matplotlib.pyplot as plt

# Define the symbol
x = symbols('x')

# Define the sigmoid expression and its derivative
#sigmoid_expr = 1 / (1 + exp(-x))
sigmoid_expr = 1 / (1 + exp(-x))
sigmoid_derivative = simplify(diff(sigmoid_expr, x))

# Convert symbolic expressions to numeric functions using lambdify
f_expr = lambdify(x, sigmoid_expr, modules='numpy')
f_derivative = lambdify(x, sigmoid_derivative, modules='numpy')

# Create array of x values
x_val = np.linspace(-10, 10, 400)

# Compute y values
y_expr = f_expr(x_val)
y_deriv = f_derivative(x_val)

# Plotting
plt.figure(figsize=(10, 4))

# Sigmoid function
plt.subplot(1, 2, 1)
plt.plot(x_val, y_expr, label=r'$\sigma(x)$')
plt.title(r'$\sigma(x)$')
plt.grid(True)

# Derivative of sigmoid
plt.subplot(1, 2, 2)
plt.plot(x_val, y_deriv, label=r'$\sigma\'(x)$', color='orange')
plt.title(r'$\frac{d\sigma(x)}{dx}$')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
from sympy import exp, diff, symbols, simplify, lambdify, Piecewise
import numpy as np
import matplotlib.pyplot as plt

# Define the symbol
x = symbols('x')

# Define the sigmoid expression and its derivative
#sigmoid_expr = 1 / (1 + exp(-x))
#tanh_expr = (exp(x)-exp(-x)) / (exp(x)+exp(-x))
tanh_expr = Piecewise((0, x <= 0), (x, x > 0))
tanh_derivative = simplify(diff(tanh_expr, x))

# Convert symbolic expressions to numeric functions using lambdify
f_expr_tanh = lambdify(x, tanh_expr, modules='numpy')
f_derivative_tanh = lambdify(x, tanh_derivative, modules='numpy')

# Create array of x values
x_val = np.linspace(-10, 10, 400)

# Compute y values
y_expr = f_expr_tanh(x_val)
y_deriv = f_derivative_tanh(x_val)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].plot(x_val, y_expr)
axs[0].set_title(r'$\tanh(x)$')
axs[0].grid(True)
axs[0].set_xlim(-10, 10)
axs[0].set_ylim(-1, 1)

# Move axes to center
axs[0].spines['left'].set_position('zero')
axs[0].spines['bottom'].set_position('zero')
axs[0].spines['right'].set_color('none')
axs[0].spines['top'].set_color('none')
axs[0].xaxis.set_ticks_position('bottom')
axs[0].yaxis.set_ticks_position('left')

# --- Plot Derivative ---
axs[1].plot(x_val, y_deriv, color='orange',linewidth=5.5)
axs[1].set_title(r'$\frac{d\tanh(x)}{dx}$')
axs[1].grid(True)
axs[1].set_xlim(-10, 10)
axs[1].set_ylim(-1, 1)

# Move axes to center
axs[1].spines['left'].set_position('zero')
axs[1].spines['bottom'].set_position('zero')
axs[1].spines['right'].set_color('none')
axs[1].spines['top'].set_color('none')
axs[1].xaxis.set_ticks_position('bottom')
axs[1].yaxis.set_ticks_position('left')

plt.tight_layout()
plt.show()
```
