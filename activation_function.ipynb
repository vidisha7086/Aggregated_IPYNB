```python

from sympy import exp, diff, symbols, simplify, lambdify
import numpy as np
import matplotlib.pyplot as plt

# Define the symbol
x = symbols('x')

# Define the sigmoid expression and its derivative
sigmoid_expr = 1 / (1 + exp(-x))
sigmoid_derivative = simplify(diff(sigmoid_expr, x))

# Convert symbolic expressions to numeric functions using lambdify
f_expr = lambdify(x, sigmoid_expr, modules='numpy')
f_derivative = lambdify(x, sigmoid_derivative, modules='numpy')

# Create array of x values
x_val = np.linspace(-10, 10, 400)

# Compute y values
y_expr = f_expr(x_val)
y_deriv = f_derivative(x_val)


# Plotting
plt.figure(figsize=(10, 4))

# Sigmoid function
plt.subplot(1, 2, 1)
plt.plot(x_val, y_expr, label=r'$\sigma(x)$')
plt.title(r'$\sigma(x)$')
plt.grid(True)

# Derivative of sigmoid
plt.subplot(1, 2, 2)
plt.plot(x_val, y_deriv, label=r'$\sigma\'(x)$', color='orange')
plt.title(r'$\frac{d\sigma(x)}{dx}$')
plt.grid(True)

plt.tight_layout()
plt.show()
```
